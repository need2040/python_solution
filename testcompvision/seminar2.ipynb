```python
import numpy as np
from skimage.io import imread, imsave, imshow
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from tqdm.auto import tqdm 
```

```python
img = imread('main.jpg')
img = img.astype(np.float32) /255
```

```python
img.dtype, img.shape
```

```python
img.min(),img.mean(), img.max()
```

```python
plt.imshow(img)
plt.show()
```

```python
img_dark = img.copy()
img_dark[..., 0]-= 50/255
img_dark = img_dark.clip(0, None)
```

```python
plt.imshow(img_dark)
plt.show() 
```

```python
img_gray = rgb2gray(img)
```
